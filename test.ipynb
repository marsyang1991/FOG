`````python
import pandas as pd
import numpy as np
import os
import test
import matplotlib.pyplot as plt                                                                                                                                 ````
`````

`````python
class fog:
    start = 0
    end = 0
    fog_start = 0
    fog_end = 0
    data = pd.DataFrame([])

    def __init__(self, start, end):
        print('A new fog item')
        self.fog_start = start
        self.fog_end = end
        self.start = self.fog_start - 64*5
        self.end = self.fog_end + 64*5

    def __repr__(self):
        return "start:{}, end:{}, data:{}".format(self.start, self.end, self.data.shape)

    def set_data(self, data):
        self.data = data
`````

`````python
def extract_fogs(data):
    start = 0
    end = 1
    in_fog = False
    fogs = []
    for index in range(0, data.shape[0]):
        label = data.loc[index, 'annotation']
        if label == 0:
            continue
        if label == 2 and not in_fog:
            start = index
            in_fog = True
        if label == 1 and in_fog:
            end = index
            new_fog = fog(start, end)
            fogs.append(new_fog)
            in_fog = False
    if len(fogs)<1:
        return []
    prev_fog = fogs[0]
    index = 1
    while index < len(fogs):
        cur_fog = fogs[index]
        print("merged")
        if cur_fog.start < prev_fog.end:
            cur_fog.start = prev_fog.end
        index += 1
        prev_fog = cur_fog
    for item in fogs:
        item.set_data(data.loc[item.start:item.end, :])
    return fogs
`````

`````python
# find file names in the direction
fdir = './dataset_fog_release/dataset/'
file_list = os.listdir(fdir)
total_feature = []
total_label = []
fogs = []
# for each file
for file_name in file_list:
    # load the data
    cur_data = test.get_data(fdir, file_name)
    fogs.extend(extract_fogs(cur_data))
print fogs
`````

`````python
def show_fog_data(fog, title):
    fog1_data = fog.data
    x = range(0, fog1_data.shape[0])
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)
    plt.title(title)
    mx_y = np.max(np.max(fog1_data.loc[:,["ankle_vertical","ankle_hori_fw","ankle_hori_lat"]]))
    mn_y = np.min(np.min(fog1_data.loc[:,["ankle_vertical","ankle_horinnnnnnnnnnnnnnnnnngbhnbnjkm,,,,mmm     _fw","ankle_hori_lat"]]))
    ax1.fill_betweenx(y=[mn_y, mx_y], x1=fog.fog_start-fog.start,x2=fog.fog_end-fog.start, facecolor='black', alpha=0.5)
    ax1.plot(x, fog1_data['ankle_vertical'], 'r')
    ax1.plot(x, fog1_data['ankle_hori_fw'], 'b')
    ax1.plot(x, fog1_data['ankle_hori_lat'], 'g')
    
    # ax1.legend(['vertical','hori_fw','hori_lat'])
    ax1.set_title("ankle")
    ax2.fill_betweenx(y=[-2000, 5000], x1=fog.fog_start-fog.start,x2=fog.fog_end-fog.start, facecolor='black', alpha=0.5)
    ax2.plot(x, fog1_data['thigh_vertical'], 'r')
    ax2.plot(x, fog1_data['thigh_hori_fw'], 'b')
    ax2.plot(x, fog1_data['thigh_hori_lat'], 'g')
    # ax2.legend(['vertical','hori_fw','hori_lat'])
    ax2.set_title("thigh")
    
    ax3.fill_betweenx(y=[-2000, 5000], x1=fog.fog_start-fog.start,x2=fog.fog_end-fog.start,facecolor='black', alpha=0.5)
    ax3.plot(x, fog1_data['trunk_vertical'], 'r')
    ax3.plot(x, fog1_data['trunk_hori_fw'], 'b')
    ax3.plot(x, fog1_data['trunk_hori_lat'], 'g')
    # ax3.legend(['vertical','hori_fw','hori_lat'])
    ax3.set_title("trunk")
    # multi = MultiCursor(fig.canvas, (ax1, ax2, ax3), color='r', lw=1)  
    plt.show()
    return fig
`````

`````python
import random
for i in range(0,20):
    index = random.randint(0,len(fogs))
    fig = show_fog_data(fogs[index],"No.{} FOG".format(index))
    fig.savefig('./img/{}_fig_{}.pdf'.format(i,index),format="pdf")
`````
